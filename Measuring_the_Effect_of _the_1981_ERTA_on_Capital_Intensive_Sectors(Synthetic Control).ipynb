```python
def add_year_suffix(df, suffix):
    df = df.copy()

    # Ensure an Industry column exists (first col is index, second is industry in your files)
    if 'Industry' not in df.columns and len(df.columns) >= 2:
        df.rename(columns={df.columns[1]: 'Industry'}, inplace=True)
        # drop first index-like column if it's not Industry
        if df.columns[0] != 'Industry':
            df.drop(columns=[df.columns[0]], inplace=True)

    # If all non-Industry columns already end with this suffix, do nothing
    other = [c for c in df.columns if c != 'Industry']
    if other and all(str(c).endswith(suffix) for c in other):
        return df

    # Rename any column that starts with a 4-digit year to "<year><suffix>"
    rename_map = {}
    for c in df.columns:
        if c == 'Industry':
            continue
        s = str(c).strip()
        m = re.match(r'^(\d{4})', s)  # grab leading year if present
        if m:
            y = m.group(1)
            target = f"{y}{suffix}"
            if s != target:
                rename_map[c] = target

    df = df.rename(columns=rename_map)

    # Keep only Industry + properly suffixed year columns
    keep = ['Industry'] + [col for col in df.columns
                           if col != 'Industry' and re.match(r'^\d{4}' + re.escape(suffix) + r'$', str(col))]
    df = df[keep]
    return df
```

```python
# --- helper: melt one metric from wide to long ---
def melt_metric(df, metric_suffix, value_name):
    # columns like "1975_net_stock" or "1975_fixed_assets"
    pattern = re.compile(r"^(\d{4})_" + re.escape(metric_suffix) + r"$")
    year_cols = [c for c in df.columns if pattern.match(c)]
    long = df.melt(
        id_vars=["Industry"],
        value_vars=year_cols,
        var_name="year_metric",
        value_name=value_name
    )
    long["Year"] = long["year_metric"].str.extract(r"^(\d{4})").astype(int)
    long = long.drop(columns=["year_metric"])
    return long

# Tag groups
healthcare = combined_df_healthcare.copy()
healthcare["Group"] = "Healthcare"

manufacturing = combined_df_manufacturing.copy()
manufacturing["Group"] = "Manufacturing"

education = combined_df_education.copy()
education["Group"] = "Education"

# Melt each metric for each group
def build_group_long(df):
    net  = melt_metric(df, "net_stock",     "net_stock")
    eqp  = melt_metric(df, "equipment",     "investment_equipment")
    fix  = melt_metric(df, "fixed_assets",  "investment_fixed_assets")
    out = (
        net
        .merge(eqp, on=["Industry","Year"], how="outer")
        .merge(fix, on=["Industry","Year"], how="outer")
    )
    # reattach Group
    out = out.merge(df[["Industry","Group"]].drop_duplicates(), on="Industry", how="left")
    return out

long_health = build_group_long(healthcare)
long_manuf  = build_group_long(manufacturing)
long_edu  = build_group_long(education)

panel = pd.concat([long_health, long_manuf,long_edu ], ignore_index=True)

# Outcome(s): investment rates (per unit of net stock)
for col in ["investment_equipment","investment_fixed_assets"]:
    panel[col + "_rate"] = panel[col] / panel["net_stock"]

# Keep analysis window if you want (1985–1990)
panel = panel[(panel["Year"]>=1975) & (panel["Year"]<=1990)].copy()

# DiD variables
panel["treated"] = (panel["Group"]=="Manufacturing").astype(int)
panel["post"]    = (panel["Year"]>=1981).astype(int)
panel["did"]     = panel["treated"]*panel["post"]

# Quick sanity check
panel.head(200)
```

```python
# --- 1) Aggregate: average (and optional CI) by Year x Group ---
def summarize_trends(panel, rate_col):
    grp = (
        panel.groupby(["Year", "Group"], as_index=False)[rate_col]
             .agg(mean="mean", count="size", std="std")
    )
    # Optional 95% CI (normal approx). Safe with count>=2; else NaN.
    grp["se"] = grp["std"] / np.sqrt(grp["count"])
    grp["ci_lo"] = grp["mean"] - 1.96*grp["se"]
    grp["ci_hi"] = grp["mean"] + 1.96*grp["se"]
    return grp

equip_trend = summarize_trends(panel, "investment_equipment_rate")
fixed_trend = summarize_trends(panel, "investment_fixed_assets_rate")

# --- 2) Plot helper with pre/post shading and ERTA line ---
def plot_trend(df, title, ylabel):
    plt.figure(figsize=(9,6))

    # Shading for pre/post
    ymin = min(df["ci_lo"].min(skipna=True), df["mean"].min())
    ymax = max(df["ci_hi"].max(skipna=True), df["mean"].max())
    plt.axvspan(1975, 1981, alpha=0.08, label="Pre-ERTA (≤1980)")
    plt.axvspan(1981, 1990, alpha=0.05, label="Post-ERTA (≥1981)")

    # ERTA vertical line
    plt.axvline(1981, linestyle="--", linewidth=1, label="ERTA (1981)")

    # Lines (and optional CI ribbons) per group
    for g in ["Manufacturing", "Healthcare", "Education"]:
        sub = df[df["Group"] == g].sort_values("Year")
        plt.plot(sub["Year"], sub["mean"], label=g, linewidth=2)
        # Confidence band if available (won't crash if NaNs)
        if sub["ci_lo"].notna().any():
            plt.fill_between(sub["Year"], sub["ci_lo"], sub["ci_hi"], alpha=0.12)

    plt.title(title)
    plt.xlabel("Year")
    plt.ylabel(ylabel)
    plt.xlim(1975, 1990)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 3) Make the two figures ---
plot_trend(
    equip_trend,
    "Trends in Equipment Investment Rates (1975–1990)",
    "Average Equipment Investment Rate"
)

plot_trend(
    fixed_trend,
    "Trends in Fixed-Asset Investment Rates (1975–1990)",
    "Average Fixed-Asset Investment Rate"
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# --- 0) Use the balanced, finite wide table you already built ---
# 'wide' has columns: Year, Manufacturing, Healthcare, Education (weighted or unweighted)
# 'treated' = "Manufacturing"
# 'donors'  = ["Healthcare", "Education"]
# pre_* and post_* already defined

# --- 1) Keep donors that vary in the pre-period (avoid std=0 explosions) ---
pre = wide[(wide["Year"]>=pre_start) & (wide["Year"]<=pre_end)].copy()
varying = [d for d in donors if pre[d].std(skipna=True) > 0]
if len(varying) == 0:
    raise ValueError("All donors are constant in pre-period. Add/clean donors.")
donors = varying

# --- 2) Build design matrices ---
y_pre = pre[treated].to_numpy(dtype=float)              # T x 1
X_pre = pre[donors].to_numpy(dtype=float)               # T x J
years_all = wide["Year"].to_numpy()
X_all = wide[donors].to_numpy(dtype=float)              # (T_total) x J
y_all = wide[treated].to_numpy(dtype=float)

# Safety: remove any remaining NaNs (should be none by your cleaning)
mask_pre = np.isfinite(y_pre) & np.isfinite(X_pre).all(axis=1)
y_pre = y_pre[mask_pre]
X_pre = X_pre[mask_pre, :]

mask_all = np.isfinite(y_all) & np.isfinite(X_all).all(axis=1)
y_all = y_all[mask_all]
X_all = X_all[mask_all, :]
years_all = years_all[mask_all]

# --- 3) Solve min_w ||y_pre - X_pre w||^2  s.t. w>=0, sum(w)=1 (SLSQP) ---
J = X_pre.shape[1]
w0 = np.ones(J) / J
bounds = [(0.0, 1.0)] * J
constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0})

def obj(w):
    r = y_pre - X_pre @ w
    return float(r @ r)

res = minimize(obj, w0, method="SLSQP", bounds=bounds, constraints=constraints,
               options={"ftol": 1e-12, "maxiter": 10_000})

if not res.success:
    raise RuntimeError(f"Synthetic weight solve failed: {res.message}")

w_hat = res.x
weights = pd.Series(w_hat, index=donors, name="weight").sort_values(ascending=False)
print("Donor weights:\n", weights)

# --- 4) Build synthetic series and GAP ---
y_syn_all = X_all @ w_hat
out = pd.DataFrame({"Year": years_all,
                    "Actual": y_all,
                    "Synthetic": y_syn_all})
out["Gap"] = out["Actual"] - out["Synthetic"]

# --- 5) Plots: Actual vs Synthetic, then Gap ---
plt.figure(figsize=(9,6))
plt.plot(out["Year"], out["Actual"], label=f"Actual {treated}", linewidth=2)
plt.plot(out["Year"], out["Synthetic"], "--", label="Synthetic", linewidth=2)
plt.axvline(1981, linestyle="--", linewidth=1)
plt.title("Synthetic Control: Actual vs Synthetic")
plt.xlabel("Year"); plt.ylabel("Investment Equipment Rate")
plt.legend(); plt.grid(alpha=.3); plt.tight_layout(); plt.show()

plt.figure(figsize=(9,5))
plt.plot(out["Year"], out["Gap"], linewidth=2, label="Gap = Actual − Synthetic")
plt.axhline(0, color="gray", linewidth=1); plt.axvline(1981, linestyle="--", linewidth=1)
plt.title("Gap Plot (Estimated ERTA Effect)"); plt.xlabel("Year"); plt.ylabel("Difference")
plt.legend(); plt.grid(alpha=.3); plt.tight_layout(); plt.show()

# --- 6) Quick diagnostics: pre-fit RMSE, RMSPE ratio ---
def rmspe(a, b):  # Root Mean Squared Prediction Error
    r = a - b
    return float(np.sqrt(np.mean(r**2)))

pre_mask = (out["Year"]>=pre_start) & (out["Year"]<=pre_end)
post_mask = (out["Year"]>=post_start) & (out["Year"]<=post_end)
pre_rmse = rmspe(out.loc[pre_mask,"Actual"].to_numpy(), out.loc[pre_mask,"Synthetic"].to_numpy())
post_rmse = rmspe(out.loc[post_mask,"Actual"].to_numpy(), out.loc[post_mask,"Synthetic"].to_numpy())
print(f"Pre-ERTA RMSE: {pre_rmse:.6f} | Post-ERTA RMSE: {post_rmse:.6f} | Ratio: {(post_rmse/pre_rmse):.2f}")
```
